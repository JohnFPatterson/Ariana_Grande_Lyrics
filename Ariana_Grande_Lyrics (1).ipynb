```python
#importing an API wrapper to pull data from Genius
import lyricsgenius
import pandas as pd
import matplotlib as plot
```

```python
#Contacting genius and identifying artist
genius = lyricsgenius.Genius("cH_wmvUnCrYSXuXTMKE8S3IcjjBPdChnkyhu942AiyC4GjXfgV4HbZu4QvC70-3w")
artist = genius.search_artist("Ariana Grande", max_songs=0)
```

```python
#Confirming we can pull lyrics
song = artist.song("thank u, next")
print(song.lyrics)
```

```python
#pulling in a list of songs from each of Grande's albums
song_list = pd.read_csv("/Users/John/Desktop/Grande_Discography.csv")
song_list.head(5)
```

```python
#making a list and a for loop that will look up the lyrics for each song in the dataframe and append them to the list
lyrics=[]
for index, row in song_list.iterrows():
    tune = row['Song']
    song = genius.search_song(tune, artist.name)
    words = song.lyrics
    lyrics.append(words)
#print(lyrics)
```

```python
#adding lyrics to dataframe
song_list["Lyrics"]=lyrics
song_list["Lyrics"]=song_list["Lyrics"].str.lower()
song_list
```

```python
#counting frequent words in lyrics
song_list["Baby Count"]=song_list['Lyrics'].str.count("baby")
song_list["Yeah Count"]=song_list['Lyrics'].str.count("yeah")
song_list["Swear Count"]=song_list['Lyrics'].str.count(" ass|damn|bitch|fuck|shit|pussy|dick")
song_list
```

```python
#exporting to csv to feed into Tableau
song_list.to_csv("/Users/John/Desktop/song_list.csv", index = False)
```

```python
#aggregating
album_list=song_list.groupby("Album").sum()
album_list
```
